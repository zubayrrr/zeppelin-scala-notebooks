```scala
// 1. Write a Scala program to get the character at the given index within a given String. Also print the length of the string.

def test(str:String, pos:Int) : Unit = {
    println(s"The character at position $pos is ${str.charAt(pos)}")
}

test("Hello", 3)
```

```scala
// 2. Write a Scala program to get the character (Unicode code point) at the given index within the String.
def test(str:String, pos:Int) : Unit = {
    println(s"Original String: $str")
    println(s"Character(unicode point) of character '${str(pos)}' is ${str.codePointAt(pos)}")
}

test("Hello", 3)
```

```scala
// 3. Write a Scala program to compare two strings lexicographically. 

def test(str1:String, str2:String) : String = {
    val res = str1.compare(str2)
    
    if (res < 0) {
        (s"'$str1' is less than '$str2'")
    } else if (res == 0) {
        (s"'$str1' is equal to '$str2'")
    } else {
        (s"'$str1' is greater than '$str2'")
    }
}

test("Hello", "Helloo")
```

```scala
// 4. Write a Scala program to concatenate a given string to the end of another string.

def test(str1:String, str2:String) : Unit = {
    val str3 = str1.concat(str2)
    println(s"The concatenated string: $str3")
}

test("Hell", "o")
```

```scala
// 5. Write a Scala program to test if a given string contains the specified sequence of char values.

def test(str:String, checkStr:String): Boolean = {
    str.contains(checkStr)
}

test("hello", "lo")
```

```scala
// 6. Write a Scala program to create a new String object with the contents of a character array. 

val arr_num = Array('1', '2', '3', '4', '5', '6')

val str = String.copyValueOf(arr_num, 0, 5)

println(str)
```

```scala
// 7. Write a Scala program to check whether a given string ends with the contents of another string. 

def test(str1:String, str2:String) : Boolean = {
    str1.endsWith(str2)
}

test("hello", "lo")
```

```scala
// 8. Write a Scala program to check whether two String objects contain the same data.

def test(str1:String, str2:String) : Boolean = {
    str1.equals(str2)
}

test("hello", "Hello")
```

```scala
// 9. Write a Scala program to compare a given string to another string, ignoring case considerations. 

def test(str1:String, str2:String) : Boolean = {
    str1.equalsIgnoreCase(str2)
}

test("hello", "Hello")
```

```scala
// 10. Write a Scala program to replace a specified character with another character.

def test(str:String, selStr:String, repStr:String): Unit = {
    val new_str = str.replace(selStr, repStr);
    println(new_str)
}

test("hello", "h", "g")
```

```scala
// 11. Write a Scala program to get a substring of a given string between two specified positions.

def test(str:String, pos1:Int, pos2:Int): Unit = {
    val new_str = str.substring(pos1, pos2);
    println(new_str)
}

test("hello", 1, 5)
```

```scala
// 12. Write a Scala program to convert all the characters to lowercase, uppercase strings.

def test(str:String): Unit = {
    val lower_str = str.toLowerCase();
    val upper_str = str.toUpperCase();
    println(lower_str)
    println(upper_str)
}

test("hello")
```

```scala
// 13. Write a Scala program to trim any leading or trailing whitespace from a given string. 

def test(str:String): Unit = {
    val new_str = str.trim();
    println(new_str)
}

test("  hello  ")
```

```scala
// 14. Write a Scala program to print after removing duplicates from a given string.

def test(str:String): String = {
    var new_str = ""
    
    for (i <- str) if (! new_str.contains(i)) new_str += i
    new_str
}

println(test("hello"))
```

```scala
// 15. Write a Scala program to find the maximum occurring character in a string. Go to the editor

def test(str:String):Char = {
    val str_set =str.toSet // will strip dupes from string
    var counter:Int = 0; var maxChar:Char = ' '
    for (i <- str_set) if (str.count(_ == i) > counter) {counter = str.count(_ == i); maxChar = i}
    maxChar
}

println(test("Hello"))
```
