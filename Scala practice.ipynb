```scala
// 2. Write a Scala program to compute the sum of the two given integer values. If the two values are the same, then return triples their sum.

def test(x:Int, y:Int) : Int = {
    if (x==y) 3 * (x + y) else x + y
}

println("Result: " + test(1, 2))
println("Result: " + test(2, 2))
```

```scala
// 3. Write a Scala program to get the absolute difference between n and 51. If n is greater than 51 return triple the absolute difference.

def test(x:Int) : Int = 
{
    val abs_diff = Math.abs(x - 51)
    if (x > 51) 3 * abs_diff else abs_diff
}

print("Result " + test(60))
print("Result " + test(30))
```

```scala
// 6. Write a Scala program to create a new string where 'if' is added to the front of a given string. If the string already begins with 'if', return the string unchanged.

def strTest(str: String) : String = {
    if (str.startsWith("if")) str else "if" + str
}

println("Result: "+ strTest("Hello"))
println("Result: "+ strTest("iftaar"))
```

```scala
// 7. Write a Scala program to remove the character in a given position of a given string. The given position will be in the range 0...string length -1 inclusive.

def strTest(str : String, n: Int): String = {
    str.take(n) + str.drop(n + 1)
}

println(strTest("Hello", 0))
println(strTest("Hello", 1))
```

```scala
// 9. Write a Scala program to create a new string which is 4 copies of the 2 front characters of a given string.If the given string length is less than 2 return the original string.

def strTest(str: String) : String = {
    if (str.length() < 2) str
    else str.substring(0, 2) * 4
}

println(strTest("Hello"))
```

```scala
// 10. Write a Scala program to create a new string with the last char added at the front and back of a given string of length 1 or more.

def strTest(str:String) : String = {
    val len = str.length()
    val lastChar = str.charAt(len - 1)
    lastChar + str + lastChar
}

println(strTest("Hello"))
```

```scala
// 20. Write a Scala program to check which number is nearest to the value 100 among two given integers. Return 0 if the two numbers are equal.

def test(x:Int, y:Int) : Int = {
    val x_diff = Math.abs(x - 100)
    val y_diff = Math.abs(y - 100)
    if (x_diff == y_diff) 0
    else if (x_diff < y_diff) x
    else y
}

println(test(95, 90))
```

```scala
// 21. Write a Scala program to check whether two given integers are in the range 40..50 inclusive, or they are both in the range 50..60 inclusive.

def test(x:Int, y:Int) : Boolean = {
     List(x, y).forall { m => m >= 40 && m <= 50 } || List(x, y).forall { n => n >= 50 && n <= 60 }
}

println(test(78, 95))
```

```scala
// 22. Write a Scala program to find the larger value from two positive integer values in the range 20..30 inclusive, or return 0 if neither is in that range.

def test(x:Int,y:Int) : Int = {
    val res = List(x,y).max
    if(res >= 20 && res <= 30) res else 0
}

println(test(25, 30))
```

```scala
// 23. Write a Scala program to check whether a given character presents in a string between 2 to 4 times.

def strTest(str:String, ichar:Char): Boolean = {
    val counts = str.count(_ == ichar)
    counts >= 2 && counts <= 4
}

println(strTest("Hello", 'l'))
println(strTest("Helo", 'l'))
```

```scala
// 25. Write a Scala program to convert the last 4 characters of a given string in upper case. If the length of the string has less than 4 then uppercase all the characters.

def strTest(str:String): String = {
    if (str.length >=4) {
        str.take(str.length() - 4) + str.drop(str.length() - 4).toUpperCase()
    }
    else str.toUpperCase()
}

println(strTest("Hello"))
println(strTest("python"))
```
